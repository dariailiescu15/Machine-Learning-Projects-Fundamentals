### 2. Analiza datelor și curățarea  

```python
print("===== Date lipsă ===== ") 
cl.isna().sum()
```

```python
# Se verifică dacă există rânduri duplicate în dataset
cl.duplicated().sum()
```

```python
# ------------------------------
# Detreminarea valorilor extreme
# ------------------------------

# Selectarea coloanelor numerice
num = cl.select_dtypes(include=['number'])

# Calcularea percentilei de 5% (Q1) și 95% (Q3)
Q1 = num.quantile(0.05)
Q3 = num.quantile(0.95)

# Calcularea IQR: diferența dintre Q3 și Q1
IQR = Q3 - Q1

# Sortarea coloanelor după IQR, descrescător
IQR.sort_values(ascending=False)
```

```python

# 1. Barplot pentru valori lipsa
# Calculam valorile lipsa pe fiecare coloană
missing = cl.isna().sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=missing.index, y=missing.values, color="skyblue") 
plt.xticks(rotation=45)
plt.title("Valori lipsă pe coloane", fontsize=14)
plt.ylabel("Numar valori lipsa")

# Adăugăm etichete pe bare
for i, v in enumerate(missing.values):
    plt.text(i, v + 5, str(v), ha='center', fontsize=10)

plt.show()

```

```python
# ----------------------------
# Eliminarea valorilor extreme
# ----------------------------

# Se păstrează doar rândurile unde toate valorile numerice se află în intervalul [Q1 - 1.5*IQR, Q3 + 1.5*IQR]
cl = cl[
    ((num >= (Q1 - 1.5 * IQR)) & (num <= (Q3 + 1.5 * IQR))).all(axis=1)
]
```

```python
#Eliminam  PassengerId  Name, Ticket, Cabin 
cl = cl.drop(['PassengerId', 'Name','Ticket','Cabin'], axis=1)
```

```python
# Eliminăm rândurile cu date lipsă
cl = cl.dropna()
```

```python
# Verificăm că s-a efectuat corect eliminarea
print("===== Date curate ===== ") 
cl.isna().sum()
```

```python
# Determinăm coeficienții de corelație
cl.corr(numeric_only=True)
```

```python
corr_cl = cl.corr(numeric_only=True).round(2)
print(corr_cl)
```

```python
# Hartă de corelare
# Crearea figurii cu dimensiunea 15x15 inch
plt.figure(figsize=(15,15))

# Crearea unui heatmap pentru matricea de corelații 'corr_vietnam'
sns.heatmap(
    data=corr_cl,   # datele care vor fi afișate în heatmap
    vmin=-1,            # valoarea minimă a culorilor (pentru scală)
    vmax=1,             # valoarea maximă a culorilor (pentru scală)
    center=0,           # centrul culorilor la 0 (pozitive vs negative)
    annot=True,         # afișează valorile numerice în celule
    cmap='cool'         # paleta de culori folosită
)

# Afișarea graficului
plt.show()
```

```python
#Eliminam  Age SibSP
cl = cl.drop(['Age','SibSp'], axis=1)
```

Am sters coloanele Age( la care initial am sters coloanele lipsa) si SibSp deoarece au corelatie foarte mica si rezulta faptul ca varsta pasagerului si numărul de frați/surori sau soți/soții la bord nu influenteaza numarul de supravietuitori.

Fare a avut valori extreme care au fost eliminate.

### 3. Preprocesarea si pregatirea pentru modelare 

```python
# Selectarea coloanelor numerice 
numerical = cl.select_dtypes(include=['number']).columns.tolist()

# Selectarea coloanele categorice (de tip obiect) 
categorical = cl.select_dtypes(include=['object']).columns.tolist()

# Afișarea rezultatelor
print('Coloane numerice:', numerical)
print('Coloane categorice:', categorical)
```

```python
# Transformarea coloanelor categorice în variabile 'dummy'
cl_final = pd.get_dummies(
    cl,
    columns=['Sex', 'Embarked'],   # coloanele categoriale relevante
    drop_first=True                # evităm colinearitatea
)

# Verificarea primelor 5 rânduri
cl_final.head()
```

```python
X = cl_final.drop('Survived',axis=1) 
y = cl_final['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=21)
```

```python
# --------------------------
# Scalarea dataset-ului folosind MinMaxScaler
# --------------------------

from sklearn.preprocessing import MinMaxScaler

# Crearea obiectului scaler
scaler = MinMaxScaler()

# Învățarea min și max pe baza întregului set X
scaler.fit(X)

# Aplicare scaler pe setul de antrenament
X_train = scaler.fit_transform(X_train)

# Aplicare scaler pe setul de test
X_test = scaler.transform(X_test)

# Transformarea rezultatul în DataFrame pentru lizibilitate, păstrând numele coloanelor
X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = pd.DataFrame(X_test, columns=X.columns)

# Afișare primelor 5 rânduri din seturile scalate
print('========== Antrenare ==========\n')
X_train.head()
```

```python
print('========== Testare ==========\n')
X_test.head()
```

```python
print(cl_final.shape)
```

```python
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

In datasetul Titanic exista cinci variabile numerice originale: Pclass, Age, SibSp, Parch, Fare si trei variabile categoriale transformate in dummy: Sex_male, Embarked_Q, Embarked_S. In total sunt opt variabile explicative. Dupa impartirea setului de date cu train_test_split in proportie de 70% pentru antrenare si 30% pentru testare, dimensiunile obtinute sunt: X_train cu 495 randuri si 8 coloane, X_test cu 213 randuri si 8 coloane, y_train cu 495 valori si y_test cu 213 valori.

### 4. Modelul de regresie logistică / Naive Bayes

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score
from sklearn.linear_model import LogisticRegression

# Setarea dimensiunii figurii și mărimea fontului
plt.rcParams['figure.figsize'] = (8,6)
plt.rcParams['font.size'] = 20

# --------------------------
# Regresie logistică
# --------------------------

logreg = LogisticRegression()

# Antrenarea modelului pe datele de training
logreg.fit(X_train, y_train)

# Prezicerea valorilor pentru setul de test
y_pred_log = logreg.predict(X_test)

# Calcularea acurateții pe setul de training
log_train = round(logreg.score(X_train, y_train) * 100, 2)

# Calcularea acurateții pe setul de test
log_accuracy = round(accuracy_score(y_pred_log, y_test) * 100, 2)

# Calcularea scorului F1
log_f1 = round(f1_score(y_pred_log, y_test) * 100, 2)

# Afișarea rezultatelor
print("Acuratețea pe setul de training :", log_train, "%")
print("Acuratețea modelului          :", log_accuracy, "%")
print("Scorul F1                     :", log_f1, "%")
print("\033[1m--------------------------------------------------------\033[0m")

# Raport de clasificare (precizie, recall, F1 pentru fiecare clasă)
print("Raport de clasificare:\n", classification_report(y_test, y_pred_log))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

# -----------------------------
# Matricea de confuzie pentru modelul de regresie logistică
# -----------------------------

plt.rcParams['figure.figsize'] = (10, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 


# Calculare matrice de confuzie
cm = confusion_matrix(y_test, y_pred_log)

# Transformarea matricei într-un DataFrame pentru vizualizare mai clară
cm_df = pd.DataFrame(cm)

# Vizualizare cu Seaborn

sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title("Matricea de Confuzie - LOG")
plt.ylabel("Valori reale")
plt.xlabel("Valori prezise")
plt.show()
```

```python
from sklearn.metrics import RocCurveDisplay

# -----------------------------
# Curba ROC pentru modelul de regresie logistică
# -----------------------------

plt.rcParams['figure.figsize'] = (14, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 

# Crearea afișajului curbei ROC pe baza estimatorului (modelului)
RocCurveDisplay.from_estimator(
    logreg,   # modelul antrenat
    X_test,   # datele de test
    y_test,   # valorile reale
)

# Titlul graficului 
plt.title("Curba ROC - Regresie Logistică")

# Afișare figură
plt.show()
```

```python
from sklearn.metrics import PrecisionRecallDisplay

# -----------------------------
# Curba Precizie-Recall pentru modelul de regresie logistică
# -----------------------------

plt.rcParams['figure.figsize'] = (14, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 

# Crearea afișajului curbei PR pe baza estimatorului (modelului)
PrecisionRecallDisplay.from_estimator(
    logreg,    # modelul antrenat
    X_test,    # datele de test
    y_test,    # valorile reale
)

# Titlul graficului 
plt.title("Curba Precizie-Recall - Regresie Logistică")

# Afișare figură
plt.show()
```

```python
from sklearn.model_selection import cross_val_score

# -----------------------------
# Scorul de validare încrucișată
# -----------------------------

# cv=10 -> împărțirea datelot în 10 fold-uri
val_score = cross_val_score(estimator=logreg, X=X_train, y=y_train, cv=10)

# Afișarea acurateții medie a modelului
print("Scorul mediu al modelului: {:.2f} %".format(val_score.mean()*100))

# Afișarea deviației standard a scorurilor pentru a vedea variația între fold-uri
print("Deviația standard: {:.2f} %".format(val_score.std()*100))
```

Modelul de regresie logistica are o acuratete de aproximativ 79% si un scor F1 de 71%, ceea ce arata ca face predictii destul de bune. Matricea de confuzie arata cate cazuri au fost prezise corect si cate au fost gresite: TN si TP sunt corecte, FP si FN sunt erori. Curba ROC arata cat de bine reuseste modelul sa diferentieze intre clase, iar un AUC de 0.87 inseamna ca modelul separa destul de bine supravietuitorii de cei care nu au supravietuit.

### 5.Modelul arborelui de decizie 

```python
# -----------------------------
# Decision Tree Classifier
# -----------------------------

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, classification_report

# Crearea modelul Decision Tree
decision = DecisionTreeClassifier()

# Antrenarea modelului pe datele de training
decision.fit(X_train, y_train)

# Prezicerea valorilor pentru setul de test
y_pred_Decision = decision.predict(X_test)

# Calcularea acurateții pe setul de training
decision_train = round(decision.score(X_train, y_train) * 100, 2)

# Calcularea acuratețeii pe setul de test
decision_accuracy = round(accuracy_score(y_pred_Decision, y_test) * 100, 2)

# Calcularea scorului F1 pe setul de test
decision_f1 = round(f1_score(y_pred_Decision, y_test) * 100, 2)

# Afișarea rezultatelor
print("Acuratețea pe setul de training :", decision_train, "%")
print("Acuratețea modelului           :", decision_accuracy, "%")
print("Scorul F1                      :", decision_f1, "%")
print("\033[1m--------------------------------------------------------\033[0m")

# Raport de clasificare (precizie, recall, F1 pentru fiecare clasă)
print("Raport de clasificare:\n", classification_report(y_test, y_pred_Decision))
```

```python
# -----------------------------
# Matricea de confuzie - Decision Tree Classifier
# -----------------------------

plt.rcParams['figure.figsize'] = (10, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 

# Calculare matrice de confuzie
cm = confusion_matrix(y_test, y_pred_Decision)

# Transformarea matriceei într-un DataFrame pentru vizualizare mai clară
cm_df = pd.DataFrame(cm)

# Vizualizare cu Seaborn

sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title("Matricea de Confuzie - Decision Tree")
plt.ylabel("Valori reale")
plt.xlabel("Valori prezise")
plt.show()
```

```python
# -----------------------------
# Curba ROC - Decision Tree Classifier
# -----------------------------

plt.rcParams['figure.figsize'] = (14, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 

# Crearea afișajului curbei ROC pe baza modelului Decision Tree
RocCurveDisplay.from_estimator(
    decision,       # modelul antrenat
    X_test,    # datele de test
    y_test     # valorile reale
)

plt.title("Curba ROC - Decision Tree")

# Afișare figură
plt.show()
```

```python
# -----------------------------
# Curba Precizie-Recall -  Decision Tree Classifier
# -----------------------------

plt.rcParams['figure.figsize'] = (14, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 

# Crearea afișajului curbei Precizie-Recall
PrecisionRecallDisplay.from_estimator(
    decision,       # modelul antrenat
    X_test,    # datele de test
    y_test     # valorile reale
)

plt.title("Curba Precizie-Recall - Decision Tree")

#Afișare figură
plt.show()
```

```python
# -------------------------------------------------------------
# Validare încrucișată (Cross-Validation) pentru Decision Tree
# -------------------------------------------------------------

# cv=10 -> împărțirea datelor în 10 fold-uri
val_score_dtc = cross_val_score(estimator=decision, X=X_train, y=y_train, cv=10)

# Afișarea scorului mediu al modelului pe toate fold-urile
print("Scorul mediu al modelului Decision Tree: {:.2f} %".format(val_score_dtc.mean()*100))

# Afișarea deviației standard pentru a evalua variația scorurilor între fold-uri
print("Deviația standard: {:.2f} %".format(val_score_dtc.std()*100))
```

Modelul Decision Tree a obtinut o acuratete de 81.22% si un scor F1 de 74.39%, ceea ce il face usor mai performant decat regresia logistica in acest caz. Matricea de confuzie arata ca modelul a prezis corect majoritatea cazurilor, cu 113 exemple negative si 61 pozitive clasificate corect. Curba ROC indica o capacitate buna de separare intre clase, cu un scor AUC de 0.80. Aceste rezultate arata ca arborele de decizie este eficient in clasificare, desi de obicei nu se foloseste independent.
